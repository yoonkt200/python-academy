## 1. 텍스트 데이터 다루기

### 1) 분석용 텍스트 데이터 수집하기

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import re

# brew 로 설치된 chromedriver의 path (Mac)
path = "/usr/local/bin/chromedriver"

# 윈도우용 크롬 웹드라이버 실행 경로 (Windows)
excutable_path = "chromedriver.exe"

# 크롤링할 페이지 주소
source_url = "https://namu.wiki/RecentChanges"

# 크롬 드라이버를 사용 (맥은 첫 줄, 윈도우는 두번째 줄 실행)
driver = webdriver.Chrome(path)
# driver = webdriver.Chrome(executable_path=excutable_path)

# 드라이버가 브라우징 할 페이지 소스를 입력
driver.get(source_url)
req = driver.page_source

# 사이트의 html 구조에 기반하여 데이터를 파싱
soup=BeautifulSoup(req, "html.parser")
contents_table = soup.find(name="table")
table_body = contents_table.find(name="tbody")
table_rows = table_body.find_all(name="tr")

# a태그의 href 속성을 리스트로 추출하여, 크롤링 할 페이지 리스트를 생성
page_url_base = "https://namu.wiki"
page_urls = []
for index in range(0, len(table_rows)):
    first_td = table_rows[index].find_all("td")[0]
    td_url = first_td.find_all("a")
    if len(td_url) > 0:
        page_url = page_url_base + td_url[0].get("href")
        if "png" not in page_url:
            page_urls.append(page_url)

# 중복 url을 제거
page_urls = list(set(page_urls))
for page in page_urls[:3]:
    print(page)
    
# 크롤링에 사용한 브라우저를 종료
driver.close()
```

```python
import pandas as pd

# 크롤링한 데이터를 데이터 프레임으로 만들기 위한 준비
columns = ["title", "category", "content_text"]
df = pd.DataFrame(columns=columns)

# 각 페이지별 '제목', '카테고리', '본문' 정보를 데이터 프레임으로
for page_url in page_urls:

    # 사이트의 html 구조에 기반하여 크롤링을 수행
    driver = webdriver.Chrome(path)  # for Mac
    # driver = webdriver.Chrome(executable_path=excutable_path)  # for Windows
    driver.get(page_url)
    req = driver.page_source
    soup = BeautifulSoup(req, "html.parser")
    contents_table = soup.find(name="article")
    title = contents_table.find_all("h1")[0]
    
    # 카테고리 정보가 없는 경우를 확인
    if len(contents_table.find_all("ul")) > 0:
        category = contents_table.find_all("ul")[0]
    else:
        category = None
        
    content_paragraphs = contents_table.find_all(name="div", attrs={"class":"wiki-paragraph"})
    content_corpus_list = []
    
    # 페이지 내 제목 정보에서 개행 문자를 제거한 뒤 추출합니다. 만약 없는 경우, 빈 문자열로 대체
    if title is not None:
        row_title = title.text.replace("\n", " ")
    else:
        row_title = ""
    
    # 페이지 내 본문 정보에서 개행 문자를 제거한 뒤 추출합니다. 만약 없는 경우, 빈 문자열로 대체
    if content_paragraphs is not None:
        for paragraphs in content_paragraphs:
            if paragraphs is not None:
                content_corpus_list.append(paragraphs.text.replace("\n", " "))
            else:
                content_corpus_list.append("")
    else:
        content_corpus_list.append("")
        
    # 페이지 내 카테고리정보에서 “분류”라는 단어와 개행 문자를 제거한 뒤 추출. 만약 없는 경우, 빈 문자열로 대체
    if category is not None:
        row_category = category.text.replace("\n", " ")
    else:
        row_category = ""
    
    # 모든 정보를 하나의 데이터 프레임에 저장
    row = [row_title, row_category, "".join(content_corpus_list)]
    series = pd.Series(row, index=df.columns)
    df = df.append(series, ignore_index=True)
    
    # 크롤링에 사용한 브라우저를 종료
    driver.close()
```

### 2) 텍스트 데이터 전처리

```python
# 텍스트 정제 함수: 한글 이외의 문자는 전부 제거
def text_cleaning(text):
    hangul = re.compile('[^ ᄀ-ᅵ가-힣]+') # 한글의 정규표현식
    result = hangul.sub('', text)
    return result

print(text_cleaning(df['content_text'][0]))
```

```python
# 각 피처마다 데이터 전처리를 적용
df['title'] = df['title'].apply(lambda x: text_cleaning(x))
df['category'] = df['category'].apply(lambda x: text_cleaning(x))
df['content_text'] = df['content_text'].apply(lambda x: text_cleaning(x))
df.head()
```

```python
# 각 피처마다 말뭉치를 생성
title_corpus = "".join(df['title'].tolist()) 
category_corpus = "".join(df['category'].tolist()) 
content_corpus = "".join(df['content_text'].tolist()) 
print(title_corpus)
```

### 3) 형태소 분석기 사용하기

```python
!pip install konlpy
```

```python
from konlpy.tag import Okt
from collections import Counter

# konlpy의 형태소 분석기 객체를 정의한다.
nouns_tagger = Okt()

# konlpy의 형태소 분석기로 명사 단위의 키워드를 추출한다.
nouns = nouns_tagger.nouns(content_corpus) 

# 추출한 명사가 각각 몇 개 등장했는지 개수를 세어준다.
count = Counter(nouns)
print(count)
```

```python
# 글자의 갯수가 1 초과인 것만 필터링함으로써, 한 글자 키워드를 제거한다
remove_char_counter = Counter({x : count[x] for x in count if len(x) > 1}) 
print(remove_char_counter)
```

```python
# 한국어 약식 불용어 사전 예시 파일. 출처 - (https://www.ranks.nl/stopwords/korean)
korean_stopwords_path = "../../dataset/korean_stopwords.txt"

# 텍스트 파일을 오픈
with open(korean_stopwords_path, encoding='utf8') as f: 
    stopwords = f.readlines()
    
# 파이썬 리스트의 형태로 변환한다.
stopwords = [x.strip() for x in stopwords] 
print(stopwords[:10])
```

```python
# 나무위키 페이지에 적용이 필요한 불용어를 추가
namu_wiki_stopwords = ['상위', '문서', '내용', '누설', 
                       '아래', '해당', '설명', '표기', 
                       '추가', ' 모든', '사용', '매우', 
                       '가장', '줄거리', '요소', '상황', 
                       '편집', '틀', '경우', '때문', '모습', 
                       '정도', '이후', '사실', '생각', 
                       '인물', '이름', '년월']

# 위 리스트의 단어들을 stopwords에 추가해준다.
for stopword in namu_wiki_stopwords:
    stopwords.append(stopword)

# 키워드 데이터에서 불용어를 제거
remove_char_counter = Counter({x : remove_char_counter[x] for x in count if x not in stopwords})
print(remove_char_counter)
```

```python
from apyori import apriori

# 장바구니 형태의 데이터(트랜잭션 데이터)를 생성
transactions = [
    ['손흥민', '시소코'], ['손흥민', '케인'],
    ['손흥민', '케인', '포체티노']
]

# 연관 분석을 수행
results = list(apriori(transactions)) 
for result in results:
    print(result)
```

```python
# 지지도 0.5, 신뢰도 0.6, 향상도 1.0 이상이면서 (손흥민, 케인)처럼 규칙의 크기가 2 이하인 규칙을 추출
list(apriori(transactions,
             min_support = 0.5, 
             min_confidence = 0.6, 
             min_lift = 1.0, 
             max_length = 2))
```

```python
# 명사 리스트를 추가할 빈 리스트를 선언한다.
nouns_list = []

# 명사 형태소 추출기를 통해 명사를 추출한다.
nouns = nouns_tagger.nouns(df['content_text'][0])

# 추출된 명사들 중에 텍스트 길이가 1이하, 그리고 불용어 사전에 있지 않은 것들만을 명사 리스트에 추가한다.
for x in nouns:
    if (len(x) > 1) and (x not in stopwords):
        nouns_list.append(x)
        
print(nouns_list)
```

```python
def get_token_list(text):
    nouns_list = []
    
    # 형태소 분석기로 명사 리스트 추출한다.
    nouns = nouns_tagger.nouns(text)
    for x in nouns:
        # 한 글자 이상이면서, 불용어 사전에 없는 키워드만 추출한다.
        if (len(x) > 1) and (x not in stopwords):
            nouns_list.append(x)
    
    return nouns_list
            
# 전체 데이터 프레임에 적용한다.
df['content_text'] = df['content_text'].apply(lambda x: get_token_list(x))
df.head()
```

```python
# 장바구니 데이터 추출
transactions = df['content_text'].tolist()

# 연관 분석 수행
results = list(apriori(transactions,
                       min_support = 0.1,
                       min_confidence = 0.25,
                       min_lift = 8.5,
                       max_length = 2))
print(len(results))
```

```python
# 데이터 프레임 형태로 정리하기 위한 빈 데이터프레임을 생성한다.
columns = ['source', 'target', 'support'] 
association_df = pd.DataFrame(columns = columns)


# 규칙의 조건절을 'source', 결과절을 'target', 지지도를 'support'라는 데이터 프레임의 피처로 변환한다.
for result in results:
    if len(result.items) == 2:
        # 분석된 규칙들을 리스트 형태로 추출한다.
        items = [x for x in result.items]
        
        # 리스트의 0번째 인덱스에 해당하는 값은 source, 1번째 인덱스에 해당하는 값은 target이다.
        # result.support는 조건의 지지도를 의미한다.
        row = [items[0], items[1], result.support]
        series = pd.Series(row, index = association_df.columns) 
        
        # 빈 데이터프레임에 시리즈 객체를 추가한다.
        association_df = association_df.append(series, ignore_index = True)
        
association_df.head()
```

### 3. 연관 키워드 시각화하기

```python
# !pip install networkx
```

```python
from collections import Counter

# 1차원의 형태로 키워드 리스트 변환한다.
flatten_array = [keyword for transaction in df['content_text'].tolist() for keyword in transaction]

# 키워드 빈도를 추출한다.
corpus = Counter(flatten_array)

# 키워드와 키워드 빈도 점수를 'node', 'nodesize'라는 데이터 프레임의 피처로 생성한다.
node_df = pd.DataFrame(corpus.items(), columns = ['node', 'nodesize'])

# 키워드가 너무 많기 때문에 샘플링을 해준다.
association_df = association_df.sample(1000)

# 샘플링된 데이터에서 등장한 node, nodesize를 리스트 형태로 정리한다.
unique_nodes = list(set(association_df['source'].unique().tolist() + association_df['target'].unique().tolist()))

# unique_nodes에 있는 노드들만 필터링한다.
node_df = node_df[node_df['node'].isin(unique_nodes)]
node_df.head()
```

```python
import matplotlib.pyplot as plt
import networkx as nx 
plt.figure(figsize = (25,25))

# networkx 그래프 객체를 생성
G = nx.Graph()

# node의 키워드 빈도수를 데이터로 하여 네트워크 그래프의 '노드' 역할을 하는 원을 생성
for index, row in node_df.iterrows(): 
    G.add_node(row['node'], nodesize=row['nodesize'])
    
# association_df 연관 분석 데이터를 기반으로 네트워크 그래프의 관계 역할을 하는 선을 생성
for index, row in association_df.iterrows(): 
    G.add_weighted_edges_from([(row['source'], row['target'], row['support'])])
    
# 그래프 디자인과 관련된 파라미터를 설정
pos = nx.spring_layout(G, k = 0.6, iterations = 50) 
sizes = [G.nodes[node]['nodesize']*25 for node in G] 
nx.draw(G, pos=pos, node_size=sizes)

# 윈도우 사용자는 'Malgun Gothic'. 그 외 OS는 OS에서 한글을 지원하는 기본 폰트를 입력
nx.draw_networkx_labels(G, pos=pos, font_family='AppleGothic', font_size=25)

# 그래프 출력
ax = plt.gca() 
plt.show()
```
